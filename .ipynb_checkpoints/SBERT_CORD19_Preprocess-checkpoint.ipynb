# Preprocess of the provided dataset

As part of the (a) task, I am going to read the papers from the given folder ```comm_use_subset ```. The contents of each json are:


- title
- authors
- abstract
- body_text
- metadata


I am going to use the body_text in order to make the SBERT model.

## __Loading__ the dataset and forming the appropriate structures without summarization

In this function I am creating the above pickle files that are formed from 1000,3000,6000 and 9000 papers:
```
- CORD19_SentenceMap_1000.pkl
- CORD19_SentenceMap_3000.pkl
- CORD19_SentenceMap_6000.pkl
- CORD19_SentenceMap_9000.pkl
```
These files contain a big dictionary that has:
```
keys:   All papers sentences 
values: A tuple like (paper_id,paper_title) 
```

I am creating this dictiony because in the QA models I made I needed quick matching between the answer (sentence) and the paper that belongs. O(1) is achived in this way.

Also, all sentences are stored as dictionary keys, and this way I feed the models with these sentences.

````python
# Opening data file
import io
from google.colab import drive
from os import listdir
from os.path import isfile, join
import json
from nltk import tokenize
import re
import progressbar
import pickle

drive.mount('/content/drive',force_remount=True)
filePath = r"/content/drive/My Drive/AI_4/2020-03-13/all_sources_metadata_2020-03-13.csv"
#desired folder: comm_use_subset
covid19_subset_path = r"/content/drive/My Drive/AI_4/comm_use_subset"
covid19_subset = pd.DataFrame(columns=['paper_id','title','abstract','body_text_asSentences','body_text'])

CORD19Dictionary = dict()
papers = 0
numofJSONs = len(listdir(covid19_subset_path))
bar = progressbar.ProgressBar(maxval=numofJSONs).start()
for f in listdir(covid19_subset_path):
  if isfile(join(covid19_subset_path, f)):
    with open(join(covid19_subset_path, f)) as json_data:
      data = json.load(json_data)
      json_content = pd.DataFrame.from_dict(data, orient='index').T
      record = []
      
      # Isolating paper id
      paper_id = json_content.loc[0,'paper_id']
      record.append(paper_id)

      # Isolating title
      title = json_content.loc[0,'metadata']['title']
      record.append(title)

      # Isolating papers abstract
      abstract = None
      for abs in json_content.loc[0,'abstract']:        
        abstract = abs['text']
      record.append(abstract)

      # Isolating papers body and concatenating to one big string
      jsonBodyText = []
      for rtext in json_content.loc[0,'body_text']:
        text = rtext['text']
        # remove hyperlinks 
        text = re.sub(r'https?:\/\/.\S+', "", text) 
        # remove hashtags 
        # only removing the hash # sign from the word 
        text = re.sub(r'[^\x00-\x7F]+',' ', text)
        text = re.sub(r'#', '', text) 
        text = re.sub(r'''(?i)\b((?:https?://|www\d{0,3}[.]|[a-z0-9.\-]+[.][a-z]{2,4}/)(?:[^\s()<>]+|\(([^\s()<>]+|(\([^\s()<>]+\)))*\))+(?:\(([^\s()<>]+|(\([^\s()<>]+\)))*\)|[^\s`!()\[\]{};:'".,<>?«»“”‘’]))''', " ", text)
        sentence =  tokenize.sent_tokenize(text)
        if len(sentence) > 3:
          jsonBodyText += sentence
      
      paperDict = dict.fromkeys(jsonBodyText, (paper_id,title))
      CORD19Dictionary.update(paperDict)
      record.append(jsonBodyText)

      body_text = ' '.join(jsonBodyText)
      record.append(body_text)

      covid19_subset.loc[len(covid19_subset)+1] = record

      if papers == 1000:
        # CORD19_Dataframe = r"/content/drive/My Drive/AI_4/CORD19_Dataframe_1000.pkl"
        # covid19_subset.to_pickle(CORD19_Dataframe)
        CORD19_Dataframe_Dict = r"/content/drive/My Drive/AI_4/CORD19_SentenceMap_1000.pkl"
        with open(CORD19_Dataframe_Dict, 'wb') as drivef:
          pickle.dump(CORD19Dictionary,drivef)
      elif papers == 3000:
        # CORD19_Dataframe = r"/content/drive/My Drive/AI_4/CORD19_Dataframe_3000.pkl"
        # covid19_subset.to_pickle(CORD19_Dataframe)
        CORD19_Dataframe_Dict = r"/content/drive/My Drive/AI_4/CORD19_SentenceMap_3000.pkl"
        with open(CORD19_Dataframe_Dict, 'wb') as drivef:
          pickle.dump(CORD19Dictionary,drivef)
      elif papers == 6000:
        # CORD19_Dataframe = r"/content/drive/My Drive/AI_4/CORD19_Dataframe_6000.pkl"
        # covid19_subset.to_pickle(CORD19_Dataframe)
        CORD19_Dataframe_Dict = r"/content/drive/My Drive/AI_4/CORD19_SentenceMap_6000.pkl"
        with open(CORD19_Dataframe_Dict, 'wb') as drivef:
          pickle.dump(CORD19Dictionary,drivef)
      
      papers+=1
      bar.update(papers)

# CORD19_Dataframe = r"/content/drive/My Drive/AI_4/CORD19_Dataframe_9000.pkl"
# covid19_subset.to_pickle(CORD19_Dataframe) 
CORD19_Dataframe_Dict = r"/content/drive/My Drive/AI_4/CORD19_SentenceMap_9000.pkl"
with open(CORD19_Dataframe_Dict, 'wb') as drivef:
  pickle.dump(CORD19Dictionary,drivef)

covid19_subset
````

### Dataset info:

````python
from statistics import mean
from collections import Counter

# Outlier review stats
titles_zerolen = Counter([len(x) for x in covid19_subset['title']])
print("Empty titles: {}".format(titles_zerolen[0]))
print('------------------------')
text_zerolen = Counter([len(x) for x in covid19_subset['body_text']])
print("Empty papers: {}".format(text_zerolen[0]))
print("Maximum paper length: {}".format(max(text_zerolen)))
print("Average paper length: ",int(mean([x for x,y in text_zerolen.items()])))
print('------------------------')
abstract_zerolen = Counter([len(x) for x in covid19_subset['abstract']])
print("Empty papers abstract : {}".format(text_zerolen[0]))
print("Maximum papers abstract length: {}".format(max(abstract_zerolen)))
print("Average papers abstract length: ",int(mean([x for x,y in abstract_zerolen.items()])))
````

I can see from the above lengths that the papers abstract is significantly smaller then the body, so I am think that the abstract might not be usefull for our question-answering model. Let now examine the content by creating some wordclouds. 

### WordCloud

Visualizing content makes much more clear the preprocess that needs to be done:



````python
import collections
from wordcloud import STOPWORDS as stopwords1,WordCloud
from nltk.corpus import stopwords as stopwords2
from nltk import word_tokenize
import string
import unicodedata
import re
import string
            
            
stopwords = set(stopwords1)
stopwords_nltk =  set(stopwords2.words('english'))
stopwords.update(stopwords_nltk)

# Function that given a list of data will produce a WordCloud
def generate_WordCloud(data,stopwords):
    textWords = []
    for x in data:
        for y in word_tokenize(x) :
            y = y.lower()
            y = WordNetLemmatizer().lemmatize(y)
            textWords.append(y)
                
    counter = collections.Counter(textWords)
    words = WordCloud(background_color ='white',stopwords=stopwords,margin=0).generate_from_frequencies(frequencies=dict(counter))

    plt.figure(figsize=(10,5))
    plt.imshow(words,interpolation='bilinear')
    plt.axis("off")
    plt.show()
````

For papers contents

````python
generate_WordCloud(covid19_subset['body_text'],stopwords)
````

For papers abstracts


````python
generate_WordCloud(covid19_subset['abstract'],stopwords)
````
